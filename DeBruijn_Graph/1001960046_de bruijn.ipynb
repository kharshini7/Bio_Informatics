```python
# 2.2 Generating K-mers 
```

```python
def generate_kmers(read, k):
  num_kmers = len(read) - k + 1
  kmers_list=[]
  for i in range(num_kmers):
      kmer = read[i:i+k]
      kmers_list.append(kmer)
  return kmers_list
```

```python
def generate_kmers_txt(inputfile,outputfile):
  input_file=open(inputfile, "r")
  output_file = open(outputfile, "w")
  for line in input_file:
    kmer=generate_kmers(line.rstrip(),2)
    output_file.write(kmer[0])
    output_file.write(" ")
    output_file.write(kmer[1])
    output_file.write("\n")
  input_file.close()
  output_file.close()
```

```python
# 2.3 Generating a De Bruijn Graph 
```

```python
def de_bruijn(file,k):
  kmers_text=open(file, 'r')
  nodes=[]
  edges=[]
  for line in kmers_text:
    nodes.append(line.rstrip().split(" "))
  print("Nodes of graph: ",nodes)
  for i in range(len(nodes)):
    edges.append((nodes[i][0],nodes[i][1]))
  print("Edges of the graph: ",edges)
  return nodes,edges
```

```python
def plot_debruijn_graph(edges, width=450, height=450):
    graph = toyplot.graph(
        [i[0] for i in edges],
        [i[1] for i in edges],
        width=width,
        height=height,
        tmarker=">", 
        vsize=25,
        vstyle={"stroke": "black", "stroke-width": 2, "fill": "none"},
        vlstyle={"font-size": "11px"},
        estyle={"stroke": "black", "stroke-width": 2},
        layout=toyplot.layout.FruchtermanReingold(edges=toyplot.layout.CurvedEdges()))
```
